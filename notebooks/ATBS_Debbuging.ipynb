```python
# box Print
def box_print(symbol, width, height):
    if len(symbol)!= 1:
        raise Exception("Symbol must be a single character string.")
    if width <= 2:
        raise Exception("Width must be greater than 2.")
    if height <= 2:
        raise Exception("Height must be greater than 2")
    
    print(symbol * width)
    for i in range(height - 2):
        print(symbol + (" " * (width - 2)) + symbol)
    print(symbol * width)

try:
    box_print("*", 4, 4)
    box_print("0", 20, 5)
    box_print("x", 1, 3)
    box_print("ZZ", 3, 3)
except Exception as err:
    print("An exception happened: " + str(err))
try:
    box_print("ZZ", 3, 3)
except Exception as err:
    print("An exception happened: " + str(err))
```

```python
# factorialLog
import logging
logging.basicConfig(level=logging.DEBUG, format="%(asctime)s - %(levelname)s - %(message)s")

# logging.disable()

logging.debug("Start of program")

def factorial(n):
    logging.debug("Start of factorial(" + str(n) + ")")
    total = 1
    for i in range(1, n+1):
        total *= i
        logging.debug("i is " + str(i) + ", total is " + str(total))
    logging.debug("End of factorial(" + str(n) + ")")
    return total

print(factorial(5))
logging.debug("End of program")
```

```python
import random, logging
logging.basicConfig(level=logging.DEBUG, format="%(asctime)s - %(levelname)s - %(message)s")
# logging.disable()

logging.debug("Start of the program")

guess = ''
while guess not in ('heads', 'tails'):
    print('Guess the coin toss! Enter heads or tails:')
    guess = input()

    logging.debug("Let us see the guess: " + str(guess))

toss = random.randint(0, 1) # 0 is tails, 1 is heads
if toss == 0:
    toss = "tails"
else:
    toss = "heads"

logging.debug("Let us see the random guess: " + str(toss) + " the guess: " + str(guess))

if toss == guess:
    print('You got it!')
else:
    print('Nope! Guess again!')
    guess = input()
if toss == guess:
    print('You got it!')
else:
    print('Nope. You are really bad at this game.')

logging.debug("What are the values of toss: " + str(toss) + " guess: " + str(guess))
```
