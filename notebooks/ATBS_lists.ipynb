```python
import random

number_of_streaks = 0

for experiment in range(1000):

    # Teil 1: 100 Münzwürfe erzeugen
    flips = []

    for i in range(100):
        if random.randint(0, 1) == 0:
            flips.append('H')
        else:
            flips.append('T')

    # Teil 2: Auf 6er-Serie prüfen
    streak_found = False

    for i in range(len(flips) - 5):
        if flips[i] == flips[i + 1] == flips[i + 2] == flips[i + 3] == flips[i + 4] == flips[i + 5]:
            streak_found = True
            break

    if streak_found:
        number_of_streaks += 1

# Ergebnis berechnen
percentage = (number_of_streaks / 10000) * 100

print('Experimente mit mindestens einer 6er-Serie:', number_of_streaks)
print('Prozentsatz:', percentage, '%')
```

```python
# Teil 1: Index & Slicing
einkauf = ["Milch", "Brot", "Eier", "Kaese", "Butter"]
print(einkauf[1])
print(einkauf[2:4])
print(einkauf[1:-2])
print(einkauf[:6])
print(einkauf[0:])
print(len(einkauf))
```

```python
# Teil 2: Methoden
einkauf1 = ["Milch", "Brot", "Eier", "Kaese", "Butter"]
einkauf1.insert(2, "Joghurt")
print(einkauf1)
pos = einkauf1.index("Kaese")
print("Kaese an Index: ", pos)
einkauf1.sort()
print("Alphabetisch sortiert: ", einkauf1)
einkauf1.reverse()
print("und jetzt umgekehrte Reihenfolge: ", einkauf1)
```

```python
# Teil 4: Multiple Assignments
a, b = "links", "rechts"
print(a, b)

c, d = "hoch", "tief"
c, d = d, c
print(c, d) # Tausch der Zuordnung ohne Umweg
```

```python
# Teil 5: Referenzen vs. Kopien
# Referenz
x = [1, 2, 3, 4]
y = x
y.append(5)
print("x = ", x) #Auch x ist verändert im Gegensatz zu Veränderung von Variablen, wo die initiale Variable unverändert bleibt
```

```python
# Flache Kopie
original = [10, 20, 30]
flach = copy.copy(original) #Neues Objekt
original.append(40)
print("original = ", original)
print("flach = ", flach) #Aenderung in original, keine Aenderung in flach, da Kopie
```

```python
# Deepcopy
matrix = [[1,2], [3, 4]]
flach2 = copy.copy(matrix) #äußere Liste kopiert
tief = copy.deepcopy(matrix) # äußere und innere Liste werden kopiert und ändern sich nicht
matrix[0].append(99)
matrix[1].append(88)
matrix.append(100)
print("flach2 = ", flach2)
print("tief = ", tief)
print("matrix = ", matrix)
```
