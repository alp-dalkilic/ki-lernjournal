```python
import random

# Zufallszahl erzeugen
secret_number = random.randint(1, 100)
print(secret_number)
print("Input a number between 1 and 100: ")

# Function: guess a number
def guess_number():
    attempts = 0
    while True:
        attempts += 1
        number = int(input(">>>"))
        if number == secret_number:
            print(f"Congratulations you hit the secret number: {number}!")
            print(attempts)
            break
        elif number < secret_number:
            print(f"Your guess: {number} is too low. Go for a higher number.")
        elif number > secret_number:
            print(f"Your guess: {number} is too high. Go for a lower number.")

        print(attempts)


guess_number()

    

```
