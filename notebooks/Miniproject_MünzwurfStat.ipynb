```python
# Münzwurf - Statistische Analyse
# Anzahl Würfe 10 - 1000 (Userangabe)
# Analyse: Gesamtzahl Würfe
# Anzahl jeweils Kopf/Zahl mit %-Angabe
# Längste Serie pro Symbo


# User-Input
import random

# Würfelergebnis
def w_ergebnis():
    global ergebnis
    ergebnis = []

    for i in range(anzahl_wuerfe):
        wurf = random.choice(["K", "Z"])
        ergebnis.append(wurf)
    return ergebnis

 # Analyse Würfelergebnis
def analyse():
    print("Gesamtzahl an Würfen: ", len(ergebnis))

    # Anzahl von K und Z und die jeweiligen Anteile
    K = 0
    Z = 0

    for i in range(len(ergebnis)):
        if ergebnis[i] == "K":
            K += 1
        else:
            Z += 1
    print("Anzahl/Anteil (%)  K: ", str(K), "/", str(round(K/anzahl_wuerfe*100,1)), "%")
    print("Anzahl/Anteil (%)  Z: ", str(Z), "/", str(round(Z/anzahl_wuerfe*100,1)), "%")

def laengste_serie():
    a_serie = 1
    l_serie_k = 0
    l_serie_z = 0

    if ergebnis[0] == "K":
        l_serie_k = 1
    else:
        l_serie_z = 1
    
    
    for i in range(1, len(ergebnis)):
        if ergebnis[i] == ergebnis[i-1]:
            a_serie += 1
        else:
            a_serie = 1

        if ergebnis[i] == "K":
            if a_serie > l_serie_k:
                l_serie_k = a_serie
        else:
            if a_serie > l_serie_z:
                l_serie_z = a_serie
    return l_serie_k, l_serie_z

            
# Hauptteil des Programmes            
print("Geben sie die gewünschte Anzahl an Würfen ein.")
while True:
    try:        
        anzahl_wuerfe = int(input("Zahl zwischen 10 und 1000: "))
        if 10 > anzahl_wuerfe or anzahl_wuerfe > 1000:
            print("Die eingegebene Zahl muss zwischen 10 und 1000 liegen")
            
        else:
            print("Gewählte Anzahl Würfe: ", str(anzahl_wuerfe))
            break
    except ValueError:
        print("Bitte geben sie eine Ganzzahl ohne Nachkommastelle ein!")
w_ergebnis()
analyse()
serie_k, serie_z = laengste_serie()
print("Längste K-Serie: ", serie_k)
print("Längste Z-Serie: ", serie_z)


```
