```python
# Einkaufslisten Manager
# Optionen anzeigen

print("""Einkaufslisten Manager zu Diensten
#######################################################
Es stehen Ihnen verschiedene Optionen zur Verfügung:
1 - Artikel hinzufügen (Die Zahl 1 eingeben)
2 - Artikel entfernen (Die Zahl 2 eingeben)
3 - Liste anzeigen (Die Zahl 3 eingeben)
4 - Liste alphabetisch sortieren (Zahl 4 eingeben)
5 - Beenden (Zahl 5 eingeben)""")


def hinzufuegen(ek_liste): # Hinzufügen von Artikeln
    item = input("Geben sie den gewünschten Artikel zum Hinzufügen ein (Umlaute als Ae Oe Ue eingeben): ")
    if item in ek_liste:
        print(f"Das Produkt {item} ist bereits auf der Einkaufsliste")
    else:
        ek_liste.append(item)

def entfernen(ek_liste): # Entfernen von Artikeln
    item = input("Welchen Artikel möchten sie entfernen? (Umlaute als Ae Oe Ue eingeben): ")
    if item not in ek_liste:
        print(f"Der Artikel {item} ist nicht auf ihrer Einkaufsliste")
    else:
        ek_liste.remove(item)

def anzeigen(ek_liste):
    print(list(enumerate(ek_liste)))

def sortieren(ek_liste):
    print(ek_liste.sort())

ek_liste = []

while True:
    try:
        eingabe = input("Geben sie die Zahl für die gewünschte Option ein: ")
        if 1 > int(eingabe) or int(eingabe) > 5:
            print("Bitte je nach gewünschter Option, die oben angegebenen Zahlen 1 bis 5 eingeben: ")
        elif eingabe == "1":
            hinzufuegen(ek_liste)
        elif eingabe == "2":
            entfernen(ek_liste)
        elif eingabe == "3":
            anzeigen(ek_liste)
        elif eingabe == "4":
            sortieren(ek_liste)
        elif eingabe == "5":
            break
    except ValueError:
        print("Bitte keine Buchstaben bei Auswahl der Option")

```
