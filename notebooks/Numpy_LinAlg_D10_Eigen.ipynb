```python
import sympy as sp
import numpy as np

# 1) Matrix from exercise 1
A = np.array([[2., 1.], 
              [1., 2.]])

# 2) Eigenvalues + Eigenvectors (numerical)
values, vectors = np.linalg.eig(A)
print("Eigenvalues: ", values) # [1., 3.] order can vary
print("Eigenvectors (Spalten): \n", vectors)

# 3) Cascade proof: Summe = Spur, Produkt = Determinante
print("Summe eigval: ", np.sum(values), " Spur eigval: ", np.trace(A))
print("Product eigval: ", np.prod(values), "det(A): ", np.linalg.det(A))

# 4) Av = Lambda . v to proof the first axis
v = vectors[:, 0] # first Eigenvector
print("v: ", v)
print("A @ v: ", A @ v)
print("lambda @ v: ", values[0] * v) # identical v is a Eigenvector

# 5) Exact with sympy
M = sp.Matrix([
    [2, 1], 
    [1, 2]
])
print("charpoly: ", M.charpoly().as_expr()) # Quadratpolynom  λ^2 - 4λ + 3
print("eigenvalues: ", M.eigenvals()) # (3, 1), (1,1)

# 6) Singularity: 0 as eigenvec (exercise 3, closes cluster D)
C = sp.Matrix([
    [2, 2],
    [1, 1]
])
print("det(C): ", C.det(), "eigenvals: ", C.eigenvals()) # 0 ==> (0,1), (3, 1)
# det(C) = 0 >>> 0 = eigenval >>> C is singular (IMT-point 10), 

```
