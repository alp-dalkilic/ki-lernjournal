```python
import numpy as np
import sympy as sp

A = np.array([[3, 1], 
              [2, 4]])

print("Matrix A = \n", A) #\n beginnt nach dem Ausdruck davor eine neue Zeile für den Output

# Determinante numerisch mit numpy
print("det(A) numpy = ", round(np.linalg.det(A), 4))

# Determinante exact mit sympy
As = sp.Matrix([[3, 1],
                [2, 4]])
print("det(A) sympy = ", As.det())

# Bedeutung det = 0 und Invertierbarkeit
print("Invertierbar?", As.det() != 0)
print("Rank(A) = ", As.rank())

#Singulärer Fall: lin. abhängige Zeilen det = 0
B = sp.Matrix([[1, 2, 3],
               [2, 4, 6], # ist gleich 2 * Zeile 1 ==> Linear abhängige zeilen
               [0, 1, 1]])
print("det(B) = ", B.det())
print("Rank von B = ", B.rank(), "( < 3 => Singulaer)")
print("Null Space = \n", B.nullspace())
```
