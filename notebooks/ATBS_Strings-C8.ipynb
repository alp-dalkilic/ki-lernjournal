```python
import pyperclip

# Copy a text into the clipboard before you run the program

text = pyperclip.paste()
lines = text.split("\n")
for i in range(len(lines)-1):
    lines[i] = "# " + lines[i]

new_text = "\n".join(lines)
#pyperclip.copy(new_text) #fügt immer nur noch ein Slash hinzu, weil er den letzten Stand immer in das Clipbord kopiert!
print(new_text)
```
