```python
import numpy as np
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
print("v1:", v1)
print("v2:", v2)
print("Dimension von v1:", v1.shape)
print("Dimension von v2:", v2.shape)
```

```python
print("v1 + v2 =", v1 + v2)
print("v1 - v2 =", v1 - v2)
```

```python
print("3 * v1 =", 3 * v1)
```

```python
lin_komb = 2 * v1 + (-1) * v2
print("2*v1 + (-1)*v2 = ", lin_komb)
```

```python
e1 = np.array([1, 0, 0])
e2 = np.array([0, 1, 0])
e3 = np.array([0, 0, 1])
print("v1 als LC:", 1 * e1 + 2 * e2 + 3 * e3)
```
