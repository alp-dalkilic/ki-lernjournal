```python
inventar = {"Seil": 1, "Fackel": 6, "Goldmuenzen": 42, "Dolch": 3, "Pfeile": 50, "Bogen": 5}

def zeige_inventar(inventar):
    gesamtzahl = 0
    for k, v in inventar.items():
        print(k, ": ", v)
        
    for v in inventar.values():
        gesamtzahl += v
    print(gesamtzahl)

def fuege_hinzu(inventar, neue_liste):
    for gegenstand in neue_liste:
        inventar.setdefault(gegenstand, 0)
        inventar[gegenstand] += 1
    return inventar

beute = ["Goldmuenzen", "Rubin", "Dolch", "Fackel", "Fackel", "Goldmuenzen", "Rubin", "Pfeile"]

print("########### INVENTAR #############################")

zeige_inventar(inventar)

print("########### BEUTE ################################")

fuege_hinzu(inventar, beute)

print(beute)

print("########### INVENTAR NACH BEUTE ##################")

zeige_inventar(inventar)

if "Rubin" in inventar:
    print("Es sind ", inventar["Rubin"], "Rubine im Inventar enthalten")
else:
    print("Es sind keine Rubine im Inventar enthalten")
```
