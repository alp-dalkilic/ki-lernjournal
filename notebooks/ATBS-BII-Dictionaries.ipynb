```python
inventar = {"Seil": 1, "Fackel": 6, "Goldmuenzen": 42, "Dolch": 3, "Pfeile": 50, "Bogen": 5}

def zeige_inventar(inventar):
    gesamtzahl = 0
    for k, v in inventar.items():
        print(k, ": ", v)
        
    for v in inventar.values():
        gesamtzahl += v
    print(gesamtzahl)

zeige_inventar(inventar)
```
