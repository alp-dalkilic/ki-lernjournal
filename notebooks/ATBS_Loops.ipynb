```python
# Swordfish ==> continue & break functions
while True:
    print("Who are you")
    name = input('>')
    if name != "Alp":
        continue
    print("Hello, Alp. What is the password? (It is a fish.)")
    password = input('>')
    if password == "Swordfish":
        break
print("Access granted.")
```

```python
# Guess the number game
import random
secret_number = random.randint(1, 20)
print("I am thinking of a number between 1 and 20 and you have 6 attempts to guess it. I just say too high or too low")

for taken_guesses in range(1, 7):
    print("Take a guess")
    guess = int(input('>>'))
    if guess > secret_number:
        print(str(guess) + " is too high")
    elif guess < secret_number:
        print(str(guess) + " is too low")
    else:
        break
if guess == secret_number:
    print(str(guess) + " was my number, very well - you got it right with " + str(taken_guesses) + " attempts")
else:
    print("Maybe next time you read my mind properly. My number was: " + str(secret_number))
```

```python
# Game: Scissors, Paper, Rock
import random, sys
print("SCISSORS, PAPER, ROCK")
wins = 0
losses = 0
ties = 0

while True: # main game loop
    print("%s Wins, %s Losses, %s Ties" % (wins, losses, ties))
    while True: # player move loop
        print("Enter your move: (r)ock (p)aper (s)cissor or (q)uit")
        player_move = input('>')
        if player_move == "q":
            sys.exit()
        if player_move == "r" or player_move == "s" or player_move == "p":
            break
        print("Type one of q, s, p, or r.")

        # display the players move
    if player_move == "r":
        print("Rock versus ")
    elif player_move == "p":
        print("Paper versus ")
    elif player_move == "s":
        print("Scissors versus ")

    # display computers move
    computer_move = random.randint(1, 3)
    if computer_move == 1:
        computer_move = "r"
        print("Rock")
    elif computer_move == 2:
        computer_move = "p"
        print("Paper")
    elif computer_move == 3:
        computer_move = "s"
        print("Scissors")

    # display and record the wins / losses / ties
    if player_move == computer_move:
        print("It is a tie")
        ties = ties + 1
    elif player_move == "r" and computer_move == "s":
        print("It is a win")
        wins = wins + 1
    elif player_move == "p" and computer_move == "r":
        print("It is a win")
        wins = wins + 1
    elif player_move == "s" and computer_move == "p":
        print("It is a win")
        wins = wins + 1
    elif computer_move == "s" and player_move == "p":
        print("It is a loss")
        losses = losses + 1
    elif computer_move == "p" and player_move == "r":
        print("It is a loss")
        losses = losses + 1
    elif computer_move == "r" and player_move == "s":
        print("It is a loss")
        losses = losses + 1
```
