```python
# yourName loop
name = ' '
while name != "Alp":
    print("Please type your name")
    name = input('>>')
print("Thank you!")
```

```python
# yourName loop with a break statement
while True:
    print("Please type your name")
    name = input('>>')
    if name == "Alp":
        break
print("Thank you!")
```

```python
# Swordfish ==> continue & break functions
while True:
    print("Who are you")
    name = input('>')
    if name != "Alp":
        continue
    print("Hello, Alp. What is the password? (It is a fish.)")
    password = input('>')
    if password == "Swordfish":
        break
print("Access granted.")
```

```python
# Übung 1: FizzBuzz
for i in range(1, 31, 1):
    if i % 3 == 0 and i % 5 == 0:
        print("Die Zahl lautet: " + str(i) + " und ist kategorisiert als: FizzBuzz")
    elif i % 3 == 0:
        print("Die Zahl lautet: " + str(i) + " und ist kategorisiert als: Fizz")
    elif i % 5 == 0:
        print("Die Zahl lautet: " + str(i) + " und ist kategorisiert als: Buzz")
    else:
        print("Die Zahl lautet: " + str(i) + " und erfüllt keine der gesetzten Bedingungen")
    
```

```python
# Übung 2:Summe und Produkt einer Zahlenreihe
summe = 0
produkt = 1 
for i in range(1, 11):
    summe = summe + i
    produkt = produkt * i
print("Summe der Zahlen 1 -10: ", summe," & Produkt der Zahlen 1 - 10: ", produkt)
```

```python
# Übung 2: Gauß mit while Schleife
total = 1
number = 0
while total <= 100:
    number = number + total
    total = total + 1
print(number)
```
