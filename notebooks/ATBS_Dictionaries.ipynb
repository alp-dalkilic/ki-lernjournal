```python
# ATBS Dictionaries

import pprint

# Teil 1: Dictionary erstellen und darauf zugreifen

person = {"name": "Alp", "alter": "54", "stadt": "Pörtschach"}
#print(person["name"])
#print(person["alter"])
# print(person["beruf"]) #Fehlermeldung, da key nicht in Dict.
person["alter"] = 55 #Value in der Dict. überschreiben
person["beruf"] = "Berater" #Neuen key ergänzen in der Dict
#print(person["beruf"], person["alter"])

# Teil 2: keys(), values(), items()

for keys in person.keys():
    print("Keys: ", keys)

for values in person.values():
    print("Values: ", values)

for keys, values in person.items():
    print(keys, " = ", values)

# Teil 3: in / not in (prüft nur schlüssel)

#print("name" in person)
#print("Alp" in person)
#print("telefon" not in person)

# Teil 4: get() - sicherer Zugriff mit Standardwert

#print(person.get("stadt", "unbekannt"))
#print(person.get("telefon", "unbekannt"))

# Teil 5:  setdefault() - setzt nur, wenn Schlüssel fehlt

person.setdefault("land", "Österreich") #Land fehlt und wird hier neu gesetzt
person.setdefault("land", "Deutschland") # Land schon gesetzt daher keine Veränderung
#print(person["land"])

# Teil 6: Zeichen zählen - das Standardzahlenmuster

text = "Hallo Welt ich bin hier und werde noch großes leisten"
zaehler = {}
for zeichen in text:
    zaehler.setdefault(zeichen, 0) #jedes Zeichen am Anfang null setzen
    zaehler[zeichen] += 1 #hochzählen
#print(zaehler)

# Teil 7: pprint - huebsche Ausgabe
#pprint.pprint(zaehler) #sortiert & Zeilenweise - gut bei großen Dicts

# Teil 8: Verschachtelte Daten (Strukturieren)
gaeste = {"Anna": {"Vorspeise": 2, "Nachtisch": 1}, "Bernd": {"Vorspeise": 0, "Nachtisch": 3},}
print(gaeste["Bernd"]["Nachtisch"]) # 3 ==> zwei Keys hintereinander

for name in gaeste:
    for gericht, anzahl in gaeste[name].items():
        print(name, " bringt ", anzahl, " x ", gericht)
```
