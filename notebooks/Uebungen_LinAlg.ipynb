```python
import numpy as np, matplotlib.pyplot as plt

v1 = np.array([5, -3])
v2 = np.array([-2, 4])
v3 = np.array([[5, -3],[3, -1]])

print(-2 * v1 + -1 * v2) # linear transformation
print(v1 + v2) # vector addition

print(v3.shape)
```

```python
v1 = np.array([5, 2])
v2 = np.array([1, 2])

points = []

for a in range(-5, 6):
    for b in range(-5, 6):
        p = a * v1 + b * v2
        points.append(p)

points = np.array(points)

plt.figure(figsize=(10, 10))

# Span-Punkte
plt.scatter(points[:, 0], points[:, 1], alpha=0.5)

# Basisvektoren
plt.quiver(0, 0, v1[0], v1[1],
           angles='xy', scale_units='xy', scale=1,
           label='v1')

plt.quiver(0, 0, v2[0], v2[1],
           angles='xy', scale_units='xy', scale=1,
           label='v2')

plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.xlim(-20, 20)
plt.ylim(-20, 20)

plt.grid()
plt.legend()
plt.show()
```
