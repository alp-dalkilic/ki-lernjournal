```python
# Lineare Transformationen als Matrizen
import numpy as np

# Teil 1: Matrix-Vektor-Produkt
# Jede innere Liste entspricht einer Zeile
A = np.array([[2, 0], 
              [0, 3]])

# Spaltenvektor als 1D-Array anlegen
v = np.array([1, 1])

# Multiplication mit @ Operator
Av = A @ v # erwartet (2, 3)
print(Av)
```

```python
# Teil 2: Geometrische Standardtansformation bauen
# e1 (1, 0) = x-Achse, e2 (0, 1) = y-Achse
# Rotation um 90° gegen den Uhrzeigersinn
R90 = np.array([[0, -1], # Rotationsmatrix
                [1, 0]])

p = np.array([3, 0]) # Vektor vor Transformation

print("Vektor nach Transformation: " + str(R90 @ p))
```
