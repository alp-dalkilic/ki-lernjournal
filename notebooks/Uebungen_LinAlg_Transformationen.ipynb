```python
# Lineare Transformationen als Matrizen
import numpy as np

# Teil 1: Matrix-Vektor-Produkt
# Jede innere Liste entspricht einer Zeile
A = np.array([[2, 0], 
              [0, 3]])

# Spaltenvektor als 1D-Array anlegen
v = np.array([1, 1])

# Multiplication mit @ Operator
Av = A @ v # erwartet (2, 3)
print(Av)
```

```python
# Teil 2: Geometrische Standardtansformation bauen
# e1 (1, 0) = x-Achse, e2 (0, 1) = y-Achse
# Rotation um 90° gegen den Uhrzeigersinn
R90 = np.array([[0, -1], # Rotationsmatrix
                [1, 0]])

p = np.array([3, 0]) # Vektor vor Transformation

print("Vektor nach Transformation: " + str(R90 @ p))
```

```python
import numpy as np
# Spiegelung X-Achse
Mx = np.array([[1, 0], 
               [0, -1]])

q = np.array([2, 3])
print("Spiegelung an der x-Achse: ", Mx @ q)
```

```python
# Skalierung um Faktor k
k = 2.5
S = k * np.eye(2) # np.eye(2) = 2x2 Einheitsmatrix
print("Skalierung x2.5: ", S @ np.array([1, 2]))
```

```python
# Teil 3: Komposition (AB), erst spiegeln und dann 90° rotieren
R90 = np.array([[0, -1], # Rotationsmatrix
                [1, 0]])
Mx = np.array([[1, 0], 
               [0, -1]])
C = R90 @ Mx
r = np.array([1, 0])
print("Composition R90 @ Mx @ [1, 0]: ", C @ r)
```

```python
# Probe mit umgekehrter Reihenfolge
R90 = np.array([[0, -1], # Rotationsmatrix
                [1, 0]])
Mx = np.array([[1, 0], 
               [0, -1]])
F = Mx @ R90
C = R90 @ Mx
r = np.array([1, 0])
print("Composition R90 @ Mx @ [1, 0]: ", C @ r, " versus umgekehrte Reihenfolge (Mx @ R90): ", F @ r)
```
