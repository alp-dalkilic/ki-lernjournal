```python
import numpy as np
import sympy as sp

# Rang, Nullity, Column-Space
# 1) Matrix as numpy-array (float so that it can be calculated with)

A = np.array([[1., 2., 3.],
              [2., 4., 6.],
              [1., 1., 1.]])
# 2) Rang = Anzahl wirklich unabhängiger Vektoren/Spalten = dim Spaltenraum
rank = np.linalg.matrix_rank(A)
print("Rank r = ", rank)

# 3) Nullity = No of columns n - rank r (Rank-Nullity-Theorem)
n = A.shape[1]
print("Nullity (n - r) = ", n - rank)

# 4) Exact null space basis mit sympy (Vektoren x mit Ax = 0)
As = sp.Matrix([[1, 2, 3],
                [2, 4, 6],
                [1, 1, 1],]) 
nullbasis = As.nullspace()
print("null space basis = ", [list(v) for v in nullbasis])

# 5) Validation: A times the null basis vector must equal the zero vector
print("A x v = ", list(As * nullbasis[0]))

# 6) Basis for the column space = pivotspalten aus der rref
rref, pivots = As.rref()
print("Pivot-columns (basis for the columns space): ", pivots) #expected [0, 1]

```
